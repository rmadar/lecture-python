```python
def my_surface(x, y):
    x0 = 5*np.sin(y)
    sigma = 5+y
    amp = (10-y)
    return amp*np.exp(-(x-x0)**2/sigma**2)
```

```python
x = np.linspace(-40, 40, 100)
y = np.linspace(0, 20, 200)
```

```python
z = my_surface(x, y)
```

```python
xx, yy = np.meshgrid(x, y)
zz = my_surface(xx, yy)
print(x.shape, y.shape, xx.shape, yy.shape)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, zz)

# Choose the default view
ax.view_init(azim=48, elev=48);
```

```python
# Rename columns names using df.rename() function
old_new_cols = {
    'Date/Time': 'date',
    'Hs': 'height',
    'Hmax': 'heightMax',
    'Tz': 'period',
    'Tp': 'energy',
    'Peak Direction': 'direction',
    'SST': 'temperature'
}

# The argument `inplace` means the current dataframe is overwritten with the change
df.rename(columns=old_new_cols, inplace=True)
colsname = list(df.columns)
data = df[colsname[1:]]
data.head()
```

```python
# Check which wave has -99 values for every variables
for c in list(data.columns):
    n = np.count_nonzero(data[c]<=-99)
    print('{}: {} wave have <=-99'.format(c, n))
```

```python
def normalize(d, c):

    # Get a numpy arrays
    v = d[c].values

    # Replace NaN by 0.0
    v[np.isnan(v)] = 0
    
    # Compute quantities
    v_mean = np.mean(v)
    v_rms = np.sqrt(np.mean((v-v_mean)**2))
    
    return (v-v_mean)/v_rms
```

```python
plt.hist(data_clean.energy, bins=np.linspace(-3, 20, 50), label='Before Normalisation');
plt.hist(normalize(data_clean,'energy'), bins=np.linspace(-3, 20, 50), label='After Normalisation');
plt.legend();
```

```python
for c in colsname[1:]:
    data_clean[c+'_norm'] = normalize(data_clean, c)
```
