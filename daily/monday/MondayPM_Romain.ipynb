```python
def sum_squared(Xs):
    '''
    Compute the sum of the squares of numbers
    stored in a list Xs.
     Xs : list of float/int/complex
    '''
    res = 0
    for x in Xs:
        res += x**2
    return res
```

1. count all the single quotes
2. replace all single quotes by `\'` (hint : check `string.replace()` doc)
3. compute the average word length in this text
4. compute the frequency of occurence of each alphabet letter
5. Text transformation:

```python
# Count the number single quote
n = 0
for l in text:
    if l=="'" : n += 1
print(n)
```

```python
# Count the number single quote - 2
characters = [l for l in text if l=="'"]
print(len(characters))
```

```python
# replace all single quotes by `\'`
text_clean = text.replace("'", "\'")
text_clean
```

```python
# Cleaning 
text_clean = text.replace(',', '').replace("'", "")

# Counting
def averaged_words_length(t):
    list_words = t.split(' ')
    mean = 0
    for w in list_words:
        mean += len(w)
    return mean / len(list_words)

# Print the results
print(f'Before cleaning: <w> = {averaged_words_length(text):.2f}')
print(f'After cleaning: <w> = {averaged_words_length(text_clean):.2f}')
```

```python
person = {'name': 'Charles', 'age': 78, 'size': 173, 'gender': 'M'}
def modify_person(p, dAge):
    p['age'] = p['age'] + dAge
    return p

def modify_person2(p, **kwargs):
    for k, v in kwargs.items():
        if k in p:
            p[k] += v
        else:
            p[k] = v
    return p

print(person)
print(modify_person(person, 3))

print(modify_person2(person, age=3, size=-10, eyes='brown'))
```

```python
addon_person = {'age':3, 'size': -10, 'eyes': 'brown', 'hair': 'long'}
```

```python
pnew = modify_person2(person, **addon_person)
print(pnew)
```
