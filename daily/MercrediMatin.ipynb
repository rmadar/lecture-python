```python
# Data
points = np.random.normal(loc=[0, 0], scale=[0.5, 0.8], size=(5000,2))
x, y = points[:, 0], points[:, 1]

points2 = np.random.normal(loc=[2, 2], scale=[0.5, 0.8], size=(5000,2))
x2, y2 = points2[:, 0], points2[:, 1]

# Plot
plt.figure(figsize=(10,6))
plt.scatter(x, y, s=100*(np.sin(x))**2, marker='o', alpha=0.1, label='Points1')
plt.scatter(x2, y2, s=100*(np.sin(x))**2, marker='o', alpha=0.1, label='Points 2')
#plt.scatter(x, y)
plt.xlim(0, 3)
plt.ylim(-3, 3)
plt.legend();
```

```python
data = np.random.normal(size=(1000, 3))
r0 = np.array([1, 4, 2])
data_trans = data + r0

xi, yi, zi = data[:,0], data[:,1], data[:,2]
xf, yf, zf = data_trans[:,0], data_trans[:,1], data_trans[:,2]
```

```python
from mpl_toolkits import mplot3d
```

```python
# Figure
plt.figure(figsize=(12,10))

ax = plt.axes(projection='3d')
ax.scatter3D(xi, yi, zi, alpha=0.4, label='before translation')
ax.scatter3D(xf, yf, zf, alpha=0.4, label='after translation')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend(frameon=False, fontsize=18);
```

```python
colName = list(df.columns)
colName.remove('Date/Time')

# Nettoyage complet
dfOK = df.copy()
for c in colName:
    isGood = dfOK[c] != -99.9
    dfOK = dfOK[isGood]
```

```python
old_new_cols = {
    'Date/Time': 'date',
    'Hs': 'height',
    'Hmax': 'heightMax',
    'Tz': 'period',
    'Tp': 'energy',
    'Peak Direction': 'direction',
    'SST': 'temperature'
}

dfOK.rename(columns=old_new_cols, inplace=True)
dfOK.head()
```

```python
colName = list(dfOK.columns)
colName.remove('date')

for c in colName:
    plt.figure()
    plt.hist(dfOK[c])
    plt.xlabel(c)
```

```python
plt.scatter(dfOK['height'], dfOK['energy'])
```

```python
def standardize(data):
    return (data - np.mean(data)) / np.std(data)

for c in colName:
    newCol = c + '_stand'
    dfOK[newCol] = standardize(dfOK[c])
    
```
