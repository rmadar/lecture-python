# scipy (curve fitting)

```python
from scipy import optimize
from scipy import stats
```

```python
x = np.random.uniform (-10, 10, size = 16)
y = np.sin(x)
dy = abs(y)**0.5
```

```python
plt.scatter(x, y)
```

```python
plt.errorbar(x,y, yerr=dy, color='black', marker='o', linestyle='none')
```

```python
x_fine = np.linspace(-10, 10, 200)
y_fine = np.sin(x_fine)
```

```python
plt.errorbar(x,y, yerr=dy, color='black', marker='o', linestyle='none', label ='data')
plt.plot(x_fine, y_fine, label='true function $sin(x)$')
plt.legend()
```

```python
# but let's pretend we didn't know it was a sin, instead let's use a polynomial to fit it
```

```python
def poly1 (x, a, b):
    y = a*x +b
    return y
```

```python
p, cov = optimize.curve_fit(poly1, x, y, sigma=dy)
```

```python
p
```

```python
poly_best = poly1(x_fine, p[0], p[1])
```

```python
p, cov**0.5
```

```python
def poly3 (x, a, b,c, d):
    y = a*x**3 +b*x**2 + c*x + d
    return y
```

```python
p, cov = optimize.curve_fit(poly3, x, y, sigma=dy)
```

```python
p, cov
```

```python
poly_best = poly3(x_fine, p[0], p[1], p[2], p[3])
```

```python
plt.errorbar(x,y, yerr=dy, color='black', marker='o', linestyle='none', label ='data')
plt.plot(x_fine, poly_best, label='poly3 fit')
plt.legend()
```
